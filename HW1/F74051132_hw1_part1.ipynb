```python
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

pd_data = pd.read_csv('./Reviews.csv')
pd_data = pd_data[:10000]
pd_data
```

```python
q1data = pd_data.copy()
ser_count = q1data.groupby(['UserId', 'ProfileName'])['Score'].count().to_frame().rename(columns={'Score':'Score count'})
ser_mean = q1data.groupby(['UserId', 'ProfileName'])['Score'].mean().to_frame().rename(columns={'Score':'Score mean'})
df_score = pd.merge(left=ser_count, right=ser_mean, on='ProfileName')
df_score.sort_values(by=['Score count'], ascending=False)[:10]
```

```python
# score = pd_data['Score'].value_counts()
q2data = q1data.set_index(['UserId', 'ProfileName'])
count = q2data.query('UserId == "A1Z54EM24Y40LL" or ProfileName == "c2"')['Score'].value_counts()
plt.bar(x=np.arange(5,0,-1), height=count, color=['purple','red','green','orange','blue'])
```

```python
import time

years = np.arange(2000, 2013)
dates = []
count = []
for n in pd_data['Time'].values:
    dates.append( time.localtime(n).tm_year )
for y in years:
    count.append( dates.count(y) )
# plt.hist(x=dates)
plt.bar(x=years, height=count)
plt.title('Review count by year')
plt.xlabel('Review year')
plt.ylabel('Review count')
```

```python
target = pd_data[['Id','HelpfulnessNumerator','HelpfulnessDenominator','Score','Time']].corr()
min_value = float(round(target.min().min(),2))
fig = sns.heatmap(target, annot=True, vmax=0.6, vmin=min_value, center=np.mean([0.6,min_value]), cmap='RdPu_r')
fig
```

```python
target = pd_data[['HelpfulnessNumerator','HelpfulnessDenominator']].rename(columns={'HelpfulnessNumerator':'n', 'HelpfulnessDenominator':'d'})
# filter out the cases with numerator > denominator
zeros = target[ (target['n'] == 0) & (target['d'] == 0) ]
ones = target[ target['n']/target['d'] == 1 ]
ratios = target[ target['n'] < target['d'] ]

neg = pd.Series(np.zeros(zeros.shape[0]) - 1)
ones = pd.Series(np.ones(ones.shape[0]))
scores = ratios['n'] / ratios['d']
scores = scores.append( ones )
scores = scores.append( neg )

plt.hist(scores.values)
```
